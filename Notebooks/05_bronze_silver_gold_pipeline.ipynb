```python
spark_df.createOrReplaceTempView("bronze_retail_sales")
```

```python
%sql
CREATE OR REPLACE TEMP VIEW silver_retail_sales AS
SELECT DISTINCT
    TransactionID,
    Date,
    Product,
    Category,
    COALESCE(Region, 'Unknown') AS Region,
    Quantity,
    Price,
    Quantity * Price AS TotalSales,
    CustomerID
FROM bronze_retail_sales
```

```python
%sql
CREATE OR REPLACE TEMP VIEW gold_monthly_revenue AS
SELECT DATE_FORMAT(Date, 'yyyy-MM') AS Month,
       SUM(TotalSales) AS Revenue
FROM silver_retail_sales
GROUP BY DATE_FORMAT(Date, 'yyyy-MM')
```
